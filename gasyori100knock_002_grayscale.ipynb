```
import cv2
img = cv2.imread("imori.jpg")
```

```
def grayscale(out):
    r = img[:, :, 2].copy()
    g = img[:, :, 1].copy()
    b = img[:, :, 0].copy()

    out[:, :, 0] = 0.2126 * r
    out[:, :, 1] = 0.7152 * g
    out[:, :, 2] = 0.0722 * b
    out = out.astype(np.uint8)
    return out
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```
img = grayscale(img)

# ファイルに出力して保存する
cv2.imwrite("out.jpg", img)
cv2_imshow(img)
```
